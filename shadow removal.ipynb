```python
import cv2
import matplotlib.pyplot as plt

def equalize_histogram(image):

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Create a CLAHE object with adaptive clip limit
    clahe = cv2.createCLAHE(clipLimit=5.0, tileGridSize=(20,20))

    # Apply CLAHE to the grayscale image
    equalized = clahe.apply(gray)

    equalized_image = cv2.cvtColor(equalized, cv2.COLOR_GRAY2BGR)

    return equalized_image

# Get user input for the image number
image_number = int(input("Enter the image number (1, 2, 3 or 4): "))

if image_number == 1:
    image_path = "/content/cctv1.JPG"
elif image_number == 2:
    image_path = "/content/cctv2.JPG"
elif image_number == 3:
    image_path = "/content/cctv3.JPG"
elif image_number == 4:
    image_path = "/content/cctv4.JPG"
else:
    print("Invalid image number. Please enter 1, 2, 3 or 4.")
    exit()

image = cv2.imread(image_path)

# Apply histogram equalization using CLAHE
equalized_image = equalize_histogram(image)

plt.imshow(cv2.cvtColor(equalized_image, cv2.COLOR_BGR2RGB))
plt.title('Equalized')
plt.axis('off')

plt.show()
```
